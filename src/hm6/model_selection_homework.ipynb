## Lựa chọn các model học máy tối ưu nhất cho dữ liệu 


```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split, learning_curve

from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# X
dataset.dropna
X = dataset.drop(['price', 'date'], axis= 1)

# Y
Y = dataset['price']

#Split dataset

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1)

#Check train and test dataset

print("Shape of train_dataset: ", X_train.shape)
print("Shape of test_dataset: ", X_test.shape)
```

```python
def cross_validation(estimator):
    _, train_scores, test_scores = learning_curve(estimator, 
                                                  X_train, Y_train, 
                                                  cv=10, 
                                                  n_jobs=-1, 
                                                  train_sizes=[1.0, ], 
                                                  scoring='neg_mean_absolute_error')
    test_scores = test_scores[0]
    mean= test_scores.mean()
    return mean
```

```python
def choosing_alpha_for_Ridge(alpha):
    best_res = -float('inf')
    best_param = None
    for param in alpha:
        text_clf = Ridge(alpha= param)
        mean = cross_validation(text_clf)
        if (mean > best_res):
            best_res = mean
            best_param = param
    return best_res, best_param


def choosing_alpha_for_Lasso(alpha):
    best_res = -float('inf')
    best_param = None
    for param in alpha:
        text_clf = Lasso(alpha= param)
        mean = cross_validation(text_clf)
        if (mean > best_res):
            best_res = mean
            best_param = param
    return best_res, best_param


def choosing_n_estimators_for_RF(n_estimators):
    best_res = -float('inf')
    best_param = None
    for param in n_estimators:
        text_clf = RandomForestRegressor(n_estimators= param, criterion='mean_absolute_error')
        mean = cross_validation(text_clf)
        if (mean > best_res):
            best_res = mean
            best_param = param
    return best_res, param


def choosing_k_neigbours_for_KNN(k_neighbours):
    best_res = -float('inf')
    best_param = None
    for param in k_neighbours:
        text_clf = KNeighborsRegressor(n_neighbors= param)
        mean = cross_validation(text_clf)
        if (mean > best_res):
            best_res = mean
            best_param = param
    return best_res, param
```

```python
alpha = [0, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
n_estimators = [10, 50]
k_neighbours = [1, 3, 5, 20]

best_res_lasso, best_alpha_for_lasso = choosing_alpha_for_Lasso(alpha)
best_res_ridge, best_alpha_for_ridge = choosing_alpha_for_Ridge(alpha)
best_res_KNN, best_k_for_KNN = choosing_k_neigbours_for_KNN(k_neighbours)
best_res_RF, best_n_for_RF = choosing_n_estimators_for_RF(n_estimators)
```

```python
print("Best alpha for Ridge: ", best_alpha_for_ridge)
print("Best alpha for Lasso: ", best_alpha_for_lasso)
print("Best k for KNN: ", best_k_for_KNN)
print("Best n for RF: ", best_n_for_RF)
rf = RandomForestRegressor(criterion='squared_error', n_estimators=best_n_for_RF)
knn = KNeighborsRegressor(n_neighbors=best_k_for_KNN)
lrg = LinearRegression()
ridge = Ridge(alpha=best_alpha_for_ridge)
lasso = Lasso(alpha=best_alpha_for_lasso)
# Huấn luyện các mô hình trên tập dữ liệu train đầy đủ
rf.fit(X_train, Y_train)
knn.fit(X_train, Y_train)
lrg.fit(X_train, Y_train)
ridge.fit(X_train, Y_train)
lasso.fit(X_train, Y_train)
```

```python
# Kết quả dự đoán trên tập test


print(f'RF: {mean_absolute_error(Y_test, rf.predict(X_test))}')
print(f'KNN: {mean_absolute_error(Y_test, knn.predict(X_test))}')
print(f'Linear Regression: {mean_absolute_error(Y_test, lrg.predict(X_test))}')
print(f'Ridge: {mean_absolute_error(Y_test, ridge.predict(X_test))}')
print(f'Lasoo: {mean_absolute_error(Y_test, lasso.predict(X_test))}')
```
