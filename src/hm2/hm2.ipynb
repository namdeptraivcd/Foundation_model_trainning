2. Cho đoạn code minh họa Biểu đồ thể hiện số bài đăng theo tháng của năm 2020. Hãy thực hiện các yêu cầu cho các câu hỏi dưới đây:

```python
df1 = newdf.groupby('month').size().reset_index(name='number of records')

n = df1['month'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df1['month'], df1['number of records'], color=c, width=.5) # value of data
for i, val in enumerate(df1['number of records'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12}) #print text float(val) as you can see the number above every columns

plt.gca().set_xticklabels(df1['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện số bài đăng theo tháng của năm 2020", fontsize=22)
plt.ylabel('number of records')
plt.show()
```

* Biểu đồ thể hiện giá nhà đất trung bình theo tháng (VND/m2)

```python
#Code ở đây
def newprice (row):
    return row['price'] / row["square"]
df = newdf.copy()
df['price_per_square'] = df.apply(newprice, axis = 1) # add a new column

df2 = df.groupby('month')["price_per_square"].mean().reset_index(name='amount of money')
df2['amount of money'] = df2['amount of money'].apply(lambda x: round(x, 2)) 
num_colors = df2['month'].unique().__len__()

all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
color = random.choices(all_colors, k = num_colors)



plt.figure(figsize=(16, 10), dpi = 80)
plt.bar(df2['month'], df2['amount of money'], color = color, width =0.5)
for i, val in enumerate(df2['amount of money'].values):
    plt.text(i, val, float(val), horizontalalignment ="center", verticalalignment = "bottom", fontdict={'fontweight':500, 'size':12})

# plt.gca().set_xticklabels(df2['month'], rotation=60, horizontalalignment = 'right')
plt.title("The graph performs the price of houses per month (VND/m2)", fontsize = 22)
plt.ylabel('amount of money')
plt.ylim(20, 35)
plt.show

```

* Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)

```python
#Gọi ý: sns.displot()
plt.figure(figsize=(13, 10), dpi = 80)
sns.distplot(df.loc[df['month'] == '12', 'price_per_square' ], color="dodgerblue", label="December", kde_kws={'linewidth':5})
sns.distplot(df.loc[df['month'] == '04', 'price_per_square'], color = "orange", label = "April", kde_kws={'linewidth':5})
plt.xlim(0, 150)
plt.title("The graph of house price density in April and December (VND/m2)", fontsize = 22)
plt.legend

```

* Số lượng bất động sản phân theo số lượng bedrooms

```python
import squarify
df1 = newdf.groupby('bedrooms').size().reset_index(name = 'counts')
labels = df1.apply(lambda x: str(x[0]) + "\n (" + str(x[1]) + ")", axis = 1)
print (labels)


num_colors = df1['bedrooms'].unique().__len__() + 1
print(num_colors)
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
colors = random.choices(all_colors, k = num_colors)
print(colors)


size = df1['counts'].values.tolist()
print(size)


plt.figure(figsize=(12, 8), dpi =80)
squarify.plot(sizes = size, label = labels, color= colors, alpha =.8)

plt.title("The number of  real estate corresponding to the number of bedrooms")
plt.axis('off') 
plt.show()


```

* Giá bất động sản trung bình theo số lượng bedrooms

```python
df3 = df.groupby('bedrooms')['price_per_square'].mean().reset_index(name = 'amount of money')
df3['amount of money'] = df3['amount of money'].apply(lambda x : round(x, 2))
df3['bedrooms'] = df3['bedrooms'].apply(lambda x : str(x))
all_colors = list(plt.cm.colors.cnames.keys())
num_colors = df3['bedrooms'].unique().__len__() + 1
random.seed(100)
colors = random.choices(all_colors, k= num_colors)

plt.figure(figsize=(16, 10))

plt.bar(df3["bedrooms"], df3['amount of money'], color = colors, width = .4)
for i, values in enumerate (df3['amount of money'].values):
    plt.text(i, values, float(values), horizontalalignment = 'center', verticalalignment = 'bottom', fontdict={'fontweight':500, 'size':12})

plt.title("House price corresponding to the number of bedrooms")
plt.ylim(20, 40)
plt.ylabel('amount of money')

```

* Giá nhà đất trung bình theo quận (VND/m2)

```python
df.columns
df4 = df.groupby('district')['price_per_square'].mean().reset_index(name = "amount of money")
df4['amount of money'] = df4['amount of money'].apply(lambda x: round (x,2))
df4 = df4.sort_values(by= 'amount of money')


all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
num_colors = df4['district'].unique().__len__() +1
colors = random.choices(all_colors, k= num_colors)

plt.figure(figsize=(16, 10), dpi= 80)
plt.bar(df4['district'], df4['amount of money'], color = colors, width= .5)
for i, values in enumerate(df4['amount of money']):
    plt.text(i, values, float(values), horizontalalignment = 'center', verticalalignment ='bottom', fontdict={'fontweight':500, 'size':13})
plt.gca().set_xticklabels(df4['district'], rotation = 60, horizontalalignment ='right')
plt.title('The price of real estate corresponding to district (VND/m2)' )
plt.ylabel('amount of money')
```

* Giá nhà đất trung bình theo nhà đầu tư (VND/m2)

```python
# Đã quen syntax nên xin phép lướt qua nhanh bằng cách cop code answer
df6 = df.groupby('investor')['price_per_square'].mean().reset_index(name='amount of money')
df6 = df6.sort_values(by = 'amount of money', ascending= False).head(10)
df6['amount of money'] = df6['amount of money'].apply(lambda x : round(x, 2))


n = df6['investor'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df6['investor'], df6['amount of money'], color=c, width=.5)
for i, val in enumerate(df6['amount of money'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df6['investor'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện giá nhà đất trung bình theo nhà đầu tư (VND/m2)", fontsize=22)
plt.ylabel('amount of money')
plt.ylim(30,70)
plt.show()
```

* Giá nhà đất trung bình theo dự án (VND/m2)

```python
df2 = df.groupby('project')['price_per_square'].mean().reset_index(name='amount of money')
df2 = df2.sort_values(by = 'amount of money', ascending= False).head(10)
# df2['amount of money'] = df2['amount of money'].apply(lambda x : int(str(x).split('.')[0]))
df2['amount of money'] = df2['amount of money'].apply(lambda x : round(x, 2))
n = df2['project'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df2['project'], df2['amount of money'], color=c, width=.5)
for i, val in enumerate(df2['amount of money'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df2['project'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện giá nhà đất trung bình theo dự án (VND/m2)", fontsize=22)
plt.ylabel('amount of money')
plt.ylim(50,80)
plt.show()
```

* Giá bất động sản phân bố theo top 10 nhà đầu tư triệu /m2

```python
# biểu đồ là top 10 nhà đầu tư, mỗi point sẽ là số tiền mà 1 dự án nhà đầu tư đổ vào

inves = df[df['investor'].notnull()]
inves = inves.groupby('investor')['price'].mean().reset_index(name='amount of money')
inves = inves.sort_values(by= "amount of money", ascending= False).head(10)
topInves = inves['investor'].unique()
df8 = df.loc[df['investor'].isin(topInves)]

plt.figure(figsize=(10, 8))
plt.scatter(df8.investor, df8.price)
plt.gca().set_xticklabels(df['investor'], rotation = 60, horizontalalignment = 'right')
plt.title('The price of real estate distributed by top 10 investors')
```

* Giá bất động sản phân bố theo dự án

```python
pr = df[df['project'].notnull()]
pr = pr.groupby('project')['price_per_square'].mean().reset_index(name = 'amount of money')
pr = pr.sort_values(by = 'amount of money', ascending= False).head(10)
protop = pr['project'].unique()

df9 = df.loc[(df['project'].isin (protop))]
plt.scatter(df9.project, df9.price_per_square)
plt.gca().set_xticklabels(df9['project'], rotation = 60, horizontalalignment = 'right')
```
